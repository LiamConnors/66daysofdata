## Plotting Dublin Bikes rental stations

Here, I work with longitude and latitude data to map locations of Dublin Bikes rental stations

```python
import pandas as pd
import folium
```

```python
location = "https://data.smartdublin.ie/dataset/33ec9fe2-4957-4e9a-ab55-c5e917c7a9ab/resource/2dec86ed-76ed-47a3-ae28-646db5c5b965/download/dublin.csv"
```

```python
bike_station_locations = pd.read_csv(location)
```

```python
bike_station_locations
```

```python
# Keeping only necessary columns
bike_station_locations = bike_station_locations[["Latitude", "Longitude", "Name"]]
```

```python
# Creating the map and adding points to it
map = folium.Map(location=[bike_station_locations.Latitude.mean(), bike_station_locations.Longitude.mean()], zoom_start=14, control_scale=True)
bike_stations = bike_station_locations[['Latitude', 'Longitude']]
bike_stations = bike_stations.values.tolist()

for point in range(0, len(bike_stations)):
    folium.Marker(bike_stations[point], popup=bike_station_locations["Name"][point]).add_to(map)

```
